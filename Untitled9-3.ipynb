````python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
````

````python
data = load_breast_cancer(as_frame=True)
X, y = data.data, data.target
X
````

````python
from sklearn.preprocessing import StandardScaler
````

````python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Превратим обратно в DataFrame для удобства
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled.head()
````

PCA ищет направления максимальной дисперсии. Если один признак измеряется в больших числах (например, area ~ сотни) и другой — в маленьких (smoothness ~ десятые доли), то первый будет доминировать. Масштабирование делает все признаки соизмеримыми (среднее = 0, стандартное отклонение = 1), чтобы PCA учитывал все признаки равноправно.

````python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
````

````python
pca = PCA().fit(X_scaled)

explained_variance = np.cumsum(pca.explained_variance_ratio_)

plt.plot(range(0, len(explained_variance)), explained_variance, marker='o')
plt.axhline(y=0.95, color='r', linestyle='--')
plt.xlabel("Число компонент")
plt.ylabel("Накопленная объясненная дисперсия")
plt.title("PCA explained variance")
plt.show()
````


```python
pca = PCA().fit(X_scaled)
```

* Создаём объект метода главных компонент (`PCA()`).
* `.fit(X_scaled)` обучает его на данных, чтобы найти новые направления (главные компоненты), которые максимизируют дисперсию.
* Пока ничего не трансформируем, просто находим структуру данных.

---

```python
explained_variance = np.cumsum(pca.explained_variance_ratio_)
```

* `pca.explained_variance_ratio_` — массив, где каждая величина показывает **долю дисперсии, объяснённой отдельной компонентой**.
  Например: `[0.44, 0.19, 0.10, …]` → первая компонента объясняет 44% информации, вторая — 19% и т.д.
* `np.cumsum(...)` — накапливает эти значения.
  Пример: `[0.44, 0.63, 0.73, …]` → первые 2 компоненты вместе объясняют 63% дисперсии, первые 3 — 73% и т.д.

---

```python
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o')
```

* Строим график зависимости «число компонент → накопленная дисперсия».
* `range(1, len(explained_variance) + 1)` даёт числа от 1 до количества компонент.
* `marker='o'` ставит кружочки на каждой точке для наглядности.

---

```python
plt.axhline(y=0.95, color='r', linestyle='--')
```

* Рисует горизонтальную красную пунктирную линию на уровне 0.95 (95%).
* Это наш «порог»: мы хотим найти минимальное число компонент, которое объясняет хотя бы 95% информации.

---

```python
plt.xlabel("Число компонент")
plt.ylabel("Накопленная объясненная дисперсия")
plt.title("PCA explained variance")
```

* Подписи к осям и заголовок графика, чтобы график был читаемым.

---

```python
plt.show()
```

* Вывод графика на экран.

---

📊 В итоге график показывает, как растёт накопленная доля объяснённой дисперсии с добавлением новых компонент.
Горизонтальная линия = наш критерий (95%).
По пересечению линии с кривой мы определяем, сколько компонент достаточно.

---

Хочешь, я нарисую, как примерно выглядит этот график именно для датасета breast cancer?


````python
n_components = np.argmax(explained_variance >= 0.95) + 1
print("Необходимое число компонент:", n_components)

pca_final = PCA(n_components=n_components)
X_pca = pca_final.fit_transform(X_scaled)

X_pca = pd.DataFrame(X_pca, columns=[f"PC{i+1}" for i in range(n_components)])
X_pca.head()
````

---

```python
n_components = np.argmax(explained_variance >= 0.95) + 1
```

* `explained_variance` — это массив накопленных значений дисперсии (например, `[0.44, 0.63, 0.73, …, 0.95, 0.96]`).
* `(explained_variance >= 0.95)` вернёт массив булевых значений (например `[False, False, …, True, True, …]`).
* `np.argmax(...)` найдёт **индекс первого `True`**, то есть первую компоненту, где накопленная дисперсия ≥ 95%.
* `+1` нужно, потому что индексы в Python начинаются с 0, а мы хотим считать компоненты с 1.
  👉 В итоге получаем минимальное количество компонент, которое объясняет ≥95% информации.

---

```python
print("Необходимое число компонент:", n_components)
```

* Просто выводим это число в консоль.

---

```python
pca_final = PCA(n_components=n_components)
```

* Создаём новый объект PCA, но теперь уже говорим: «Оставь только `n_components` компонент».
* То есть мы не просто анализируем данные, а реально уменьшаем их размерность.

---

```python
X_pca = pca_final.fit_transform(X_scaled)
```

* `fit_transform` обучает PCA на данных и сразу применяет преобразование.
* Результат — новая матрица признаков: вместо 30 исходных столбцов теперь, например, 10 (в зависимости от того, что получилось в `n_components`).
* Эти новые признаки — **главные компоненты (PC1, PC2, …)**, линейные комбинации исходных.

---

```python
X_pca = pd.DataFrame(X_pca, columns=[f"PC{i+1}" for i in range(n_components)])
```

* Превращаем массив NumPy обратно в DataFrame для удобства.
* `columns=[f"PC{i+1}" for i in range(n_components)]` создаёт список имён: `["PC1", "PC2", "PC3", …]`.

---

```python
X_pca.head()
```

* Показываем первые строки нового набора данных (уже сжатого PCA).

---

📌 **Итог**:

* Да, можно было посчитать точки на графике, это ручной способ.
* Но код автоматизирует это и даёт точное число компонент.
* После этого мы реально уменьшаем размерность данных и получаем новый датасет, где меньше признаков, но сохранилось ≥95% информации.

---


````python
from sklearn.model_selection import train_test_split

print("Баланс классов:\n", y.value_counts(normalize=True))

# Разделение PCA-данных
X_train_pca, X_test_pca, y_train, y_test = train_test_split(
    X_pca, y, test_size=0.15, stratify=y, random_state=42
)

# Разделение исходных данных (масштабированных)
X_train_scaled, X_test_scaled, _, _ = train_test_split(
    X_scaled, y, test_size=0.15, stratify=y, random_state=42
)
````

Давай разберём этот кусочек кода по шагам 🔎

---

```python
from sklearn.model_selection import train_test_split
```

* Импортируем функцию для разделения данных на обучающую и тестовую выборки.

---

```python
print("Баланс классов:\n", y.value_counts(normalize=True))
```

* `y` — это целевая переменная (0 = malignant, 1 = benign).
* `value_counts(normalize=True)` посчитает **долю** каждого класса.
  Например:

```
1    0.627417
0    0.372583
```

То есть примерно **63% здоровых / 37% больных**.
Это важно, потому что если бы классы были очень несбалансированы, пришлось бы применять специальные методы.

---

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(
    X_pca, y, test_size=0.15, stratify=y, random_state=42
)
```

* `X_pca` — это **преобразованные PCA данные**.
* `test_size=0.15` → 15% данных уйдёт в тест, 85% — в тренировку.
* `stratify=y` → разбиение будет **стратифицированным**, то есть в train и test сохранится тот же баланс классов, что и во всём наборе (63/37).
* `random_state=42` → фиксируем «случайность» для воспроизводимости.
* В результате:

  * `X_train_pca`, `y_train` → обучающая выборка.
  * `X_test_pca`, `y_test` → тестовая выборка.

---

```python
X_train_scaled, X_test_scaled, _, _ = train_test_split(
    X_scaled, y, test_size=0.15, stratify=y, random_state=42
)
```

* То же самое, только теперь работаем с **исходными масштабированными признаками** (до PCA).
* Зачем? 👉 чтобы потом сравнить результаты модели на полном наборе признаков (30 штук) и на PCA-признаках (например, 10 штук).
* Мы игнорируем `y_train` и `y_test` тут (ставим `_`), потому что они у нас уже есть из предыдущего разбиения — и они будут идентичными (так как `random_state` один и тот же).

---

📌 **Итог**:

* Мы проверили баланс классов (он умеренно смещён, но не критично).
* Сделали два разбиения:

  1. На данных после PCA.
  2. На исходных данных (после масштабирования).
* Это позволит позже обучить **две модели логистической регрессии** и сравнить качество.

---

Хочешь, я покажу тебе конкретные значения баланса классов для breast cancer датасета и размеры train/test после такого разбиения?


````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score

# Модель на PCA-признаках
clf_pca = LogisticRegression(max_iter=5000)
clf_pca.fit(X_train_pca, y_train)
y_pred_pca = clf_pca.predict(X_test_pca)
f1_pca = f1_score(y_test, y_pred_pca)

# Модель на исходных масштабированных данных
clf_orig = LogisticRegression(max_iter=5000)
clf_orig.fit(X_train_scaled, y_train)
y_pred_orig = clf_orig.predict(X_test_scaled)
f1_orig = f1_score(y_test, y_pred_orig)

print("F1-score на PCA:", f1_pca)
print("F1-score на исходных данных:", f1_orig)
````


---

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
```

* Импортируем **логистическую регрессию** — классический алгоритм классификации.
* Импортируем метрику **f1-score**, которая учитывает и точность (precision), и полноту (recall). Для медицинских задач это особенно важно, потому что нам важно не пропустить больных пациентов.
